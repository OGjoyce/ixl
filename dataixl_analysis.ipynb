```python
import pandas
import numpy
#IXL
```

```python
#import data
data = pandas.read_csv("data.csv")
```

```python
data
```

```python
#Remove NA where all data is missing
data.dropna(axis='rows',how='all')
```

```python
data.fillna('(en blanco)', inplace=True)
```

```python
#Get all the possible classes
seniority = data['Seniority Level'].value_counts(dropna=False)
position = data['Position'].value_counts(dropna=False)
related = data['Is Job Related to Innovation'].value_counts(dropna=False)
department = data['Department'].value_counts(dropna=False)
industry = data['Industry'].value_counts(dropna=False)
employee = data['Employee'].value_counts(dropna=False)
country = data['Country'].value_counts(dropna=False)
ternure = data['Tenure'].value_counts(dropna=False)
```

```python
seniority
```

```python
dfseniority = pandas.DataFrame(seniority)
dfposition = pandas.DataFrame(position)
dfrelated = pandas.DataFrame(related)
dfdepartment = pandas.DataFrame(department)
dfindustry = pandas.DataFrame(industry)
dfemployee = pandas.DataFrame(employee)
dfcountry = pandas.DataFrame(country)
dfternure = pandas.DataFrame(ternure)
```

```python
#Convert to csv
dfseniority.to_csv('dfseniority.csv')
dfposition.to_csv('dfposition.csv')
dfrelated.to_csv('dfrelated.csv')
dfdepartment.to_csv('dfdepartment.csv')
dfindustry.to_csv('dfindustry.csv')
dfemployee.to_csv('dfemployee.csv')
dfcountry.to_csv('dfcountry.csv') 
dfternure.to_csv('dfternure.csv')
```
